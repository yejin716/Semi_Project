```python
import folium
from geopy.geocoders import Nominatim
import pandas as pd
import time

df = df[df['도로명주소'].str.contains('서울')]

# Folium 맵 객체 생성
m = folium.Map(location=[37.5665, 126.978], zoom_start=12)  # 서울의 중심 좌표

# 주소를 위도와 경도로 변환
geolocator = Nominatim(user_agent="geo_app")

# 각 구역별 색상 설정
colors = {
    '강남구': 'blue',
    '서초구': 'green',
    '송파구': 'orange',
    '마포구': 'red',
    '기타': 'gray'  # 나머지 구역 색상
}

# 각 구역별 마커 추가
for address in df['도로명주소']:
    attempts = 3  # 최대 재시도 횟수
    for attempt in range(attempts):
        try:
            location = geolocator.geocode(address)
            if location:  # 위치가 성공적으로 찾은 경우
                # 구역을 확인하여 색상 결정
                area_found = False
                for area, color in colors.items():
                    if area in address:
                        folium.Marker(
                            [location.latitude, location.longitude],
                            popup=address,
                            tooltip=f"{area}",
                            icon=folium.Icon(color=color)  # 각 구역에 맞는 색상
                        ).add_to(m)
                        area_found = True
                        break  # 마커 추가 후 다음 주소로 넘어감
                if not area_found:  # 지정된 구역에 해당하지 않는 경우
                    folium.Marker(
                        [location.latitude, location.longitude],
                        popup=address,
                        tooltip="기타 구역",
                        icon=folium.Icon(color=colors['기타'])  # 기타 구역 색상
                    ).add_to(m)
            break  # 위치를 찾으면 재시도 종료
        except Exception as e:
            print(f"Error occurred for {address}: {e}")
            time.sleep(1)  # 1초 대기 후 재시도

# 지도 출력
m.save('seoul_map_with_others.html')  # HTML 파일로 저장
m  # Jupyter Notebook에서 지도를 표시하려면 이렇게 실행
```
