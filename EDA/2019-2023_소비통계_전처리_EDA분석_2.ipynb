```python
#평균값

import matplotlib.pyplot as plt
import seaborn as sns


def visualize_expenses_by_month_and_category(df, month, category, name):
    # 해당 조건에 맞는 데이터 필터링
    filtered_data = df[df['조사연월'] == month][category]
    
    # filtered_data의 평균값 계산
    filtered_data_mean = filtered_data.mean()

    # 막대 그래프(bar plot) 그리기
    plt.figure(figsize=(10, 8))
    bars = plt.bar(filtered_data.columns, filtered_data_mean)
    
    # 각 막대 위에 비용 표시
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 100, int(bar.get_height()), ha='center', fontsize=10)
    plt.title(f"{month}년도 {name}의 총합 비율",fontsize=20)
    plt.xlabel(f"{name}", fontsize=10)
    plt.ylabel(f"{name}의 비용",fontsize=10)
    plt.xticks(fontsize=10,rotation=90)
    plt.yticks(fontsize=10)
    
    # 파이 차트를 위한 데이터 필터링
    filtered_data_for_pie = filtered_data.T[filtered_data.values[0] != 0].T
    
    # 파이 차트 그리기
    plt.figure(figsize=(15, 10))
    plt.pie(filtered_data_for_pie.values[0], labels=filtered_data_for_pie.columns, autopct='%1.1f%%', startangle=140,labeldistance=1.0, textprops={'fontsize': 10})
    plt.title(f'{month}년도 {name}의 비율')
    plt.axis('equal')

    
    plt.tight_layout()
    plt.show()
```
