```python
single_20_expend = data[(data['가구원수'] == 1) &
                     (data['연령대'] == '20대')][["소비지출",'교통비', '주거수도광열_실제주거비구입', '음식숙박비', '보험료', '식료품_육류구입', '보건_외래의료서비스이용', '가정용품가사서비스_가전가정용기기구입']]
single_20_expend.head(2)
```

```python
from scipy import stats

# 정규성 검정 (Shapiro-Wilk test)
def shapiro_test(data):
    for column in data.columns[1:]:
        stat, p_value = stats.shapiro(data[column])
        print(f"{column}: Shapiro-Wilk Test - Statistic: {stat}, p-value: {p_value}")
        if p_value < 0.05:
            print(f"{column}: 데이터는 정규성을 만족하지 않습니다.")
            print('='*100)
        else:
            print(f"{column}: 데이터는 정규성을 만족합니다.")
            print('='*100)

# 등분산성 검정 (Levene's test)
def levene_test(data):
    stat, p_value = stats.levene(*[data[column] for column in data.columns[1:]])
    print(f"Levene's Test - Statistic: {stat}, p-value: {p_value}")
    if p_value > 0.05:
        print("데이터는 등분산성을 만족합니다.")
        print('='*100)
    else:
        print("데이터는 등분산성을 만족하지 않습니다.")
        print('='*100)

```

```python
# 정규성 검정
shapiro_test(single_20_expend)
```

```python
# 등분산성 검정
levene_test(single_20_expend)
```

```python
# Wilcoxon 부호 순위 검정 이 방법은 모집단의 분포에 대한 가정을 필요로하지 않으며, 특히 정규성을 가정할 수 없거나 이상치가 있는 경우에 유용합니다. (비모수적 검정)
def wilcoxon_test(data):
    for column in data.columns[1:]:
        stat, p_value = stats.wilcoxon(data['소비지출'], data[column])
        print(f"Wilcoxon Rank Sum Test for {column}: Statistic: {stat}, p-value: {p_value}")
        if p_value > 0.05:
            print(f"{column}과 소비지출 간의 차이는 통계적으로 유의하지 않습니다.")
            print('='*100)
        else:
            print(f"{column}과 소비지출 간의 차이는 통계적으로 유의합니다.")
            print('='*100)

from scipy.stats import levene, mannwhitneyu

def levene_test(data):
    stat, p_value = levene(*[data[col] for col in data.columns])
    print(f"Levene's Test - Statistic: {stat}, p-value: {p_value}")
    if p_value > 0.05:
        print("등분산성을 만족합니다.")
        print('='*100)
    else:
        print("등분산성을 만족하지 않습니다.")
        print('='*100)

def mann_whitney_u_test(data):
    for col in data.columns[1:]:
        stat, p_value = mannwhitneyu(data['소비지출'], data[col])
        print(f"Mann-Whitney U Test for {col}: Statistic: {stat}, p-value: {p_value}")
        if p_value > 0.05:
            print(f"{col}과 소비지출 간의 차이는 통계적으로 유의하지 않습니다.")
            print('='*100)
        else:
            print(f"{col}과 소비지출 간의 차이는 통계적으로 유의합니다.")
            print('='*100)


```

```python
# Wilcoxon 부호 순위 검정 비모수적 검정
wilcoxon_test(single_20_expend)
```

```python
# Mann-Whitney U 검정 수행 이 방법은 두 그룹의 데이터가 정규 분포를 따르지 않거나 등분산성을 가정할 수 없는 경우에 유용하게 사용됩니다. (비모수적 검정)
mann_whitney_u_test(single_20_expend)
```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# 정규성 검정을 위한 QQ 플롯 그리기
def qq_plot(data):
    # QQ plot 그리기
    stats.probplot(data, dist="norm", plot=plt)
    plt.title("Q-Q Plot")
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Sample Quantiles")
    plt.show()

# 소비지출에 대한 QQ plot 그리기
qq_plot(single_20_expend['소비지출'])
```
